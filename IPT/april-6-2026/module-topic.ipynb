# Python Modules and Imports

This notebook demonstrates how to work with Python modules, including:
- Importing modules
- Using module functions
- Accessing module variables
- Using module aliases
- Importing specific objects from modules
- Working with standard library modules like `platform`

Each section can be run independently!

````python
# Create calc module with basic functions
import sys
from types import ModuleType

# Create calc module
calc = ModuleType('calc')

# Basic functions
def add(x, y):
    """Add two numbers"""
    return x + y

def subtract(x, y):
    """Subtract two numbers"""
    return x - y

# 5 Additional functions that utilize add and subtract
def multiply(x, y):
    """Multiply using repeated addition"""
    result = 0
    for _ in range(y):
        result = add(result, x)
    return result

def divide_integer(x, y):
    """Integer division using repeated subtraction"""
    if y == 0:
        return "Error: Division by zero"
    result = 0
    temp = x
    while temp >= y:
        temp = subtract(temp, y)
        result = add(result, 1)
    return result

def power(x, y):
    """Calculate x to the power of y using multiply"""
    if y == 0:
        return 1
    result = x
    for _ in range(y - 1):
        result = multiply(result, x)
    return result

def sum_list(numbers):
    """Sum all numbers in a list using add"""
    result = 0
    for num in numbers:
        result = add(result, num)
    return result

def difference_sum(num1, num2, num3):
    """Calculate (num1 + num2) - num3"""
    sum_val = add(num1, num2)
    result = subtract(sum_val, num3)
    return result

# Add all functions to calc module
calc.add = add
calc.subtract = subtract
calc.multiply = multiply
calc.divide_integer = divide_integer
calc.power = power
calc.sum_list = sum_list
calc.difference_sum = difference_sum

# Register in sys.modules
sys.modules['calc'] = calc

print("✓ calc module created successfully!")
print("  Functions: add, subtract, multiply, divide_integer, power, sum_list, difference_sum")
````

## Section 3: Accessing Dictionary Variables from Modules

You can also access variables (like dictionaries) from imported modules. Use bracket notation to access specific keys.

````python
# Access the person1 dictionary from mymodule
import mymodule

# Print the entire dictionary
print("=== Full dictionary ===")
print(mymodule.person1)

# Access specific keys from the dictionary
print("\n=== Accessing specific keys ===")
age = mymodule.person1["age"]  # 'age' is a key
print(f"Age: {age}")

name = mymodule.person1["name"]
print(f"Name: {name}")

country = mymodule.person1["country"]
print(f"Country: {country}")

# Store in a variable and print
print("\n=== Store age in variable ===")
a = mymodule.person1["age"]
print(f"Stored age value: {a}")
````

## Section 4: Using Module Aliases

Module aliases allow you to give a module a shorter name using the `import module as alias` syntax. This is useful for long module names.

````python
# Import mymodule with an alias
import mymodule as mx  # 'mx' is the alias for mymodule

print("=== Using module alias 'mx' ===")
print("Instead of 'mymodule.person1', we use 'mx.person1'")

# Access person1 using the alias
a = mx.person1["age"]
print(f"Age: {a}")

# Call functions using the alias
print("\n=== Calling function with alias ===")
mx.greeting("Charlie")

# Show all person information using alias
print("\n=== Full person information ===")
for key, value in mx.person1.items():
    print(f"{key}: {value}")
````

## Summary: Key Concepts

### Import Methods:

| Syntax | Usage | Example |
|--------|-------|---------|
| `import module` | Import entire module | `import mymodule` |
| `import module as alias` | Import with shorter name | `import mymodule as mx` |
| `from module import object` | Import specific items | `from mymodule import person1` |
| `from module import *` | Import all items (not recommended) | `from mymodule import *` |

### Key Points:

1. **Modules** - Collections of functions, variables, and classes
2. **Import entire module** - Access items with `module.item`
3. **Module aliases** - Use `as` keyword for shorter names
4. **Selective imports** - Import only what you need with `from...import`
5. **Standard library** - Python comes with modules like `platform`, `os`, `sys`, etc.
6. **Dictionary access** - Use bracket notation `dict["key"]` to access values
7. **Functions** - Can be called after importing with `function(args)`

### Different Import Patterns:

```python
# Pattern 1: Import entire module
import mymodule
mymodule.greeting("John")

# Pattern 2: Use alias
import mymodule as mx
mx.greeting("John")

# Pattern 3: Import specific items
from mymodule import greeting, person1
greeting("John")
print(person1["age"])

# Pattern 4: Import multiple items
from mymodule import greeting, person1, greeting as greet
```

## Section 9: Creating calculator_app Module

Now let's create a `calculator_app` module that utilizes the calc module functions. This module will have high-level operations that call our 5 functions.

````python
# Create calculator_app module that uses calc module
import sys
from types import ModuleType

# Import calc functions
import calc

# Create calculator_app module
calculator_app = ModuleType('calculator_app')

# High-level functions that use calc module
def square(x):
    """Calculate x squared using power function"""
    return calc.power(x, 2)

def cube(x):
    """Calculate x cubed using power function"""
    return calc.power(x, 3)

def area_rectangle(length, width):
    """Calculate rectangle area using multiply"""
    return calc.multiply(length, width)

def average(numbers):
    """Calculate average of numbers"""
    total = calc.sum_list(numbers)
    count = len(numbers)
    return calc.divide_integer(total, count)

def factorial(n):
    """Calculate factorial using multiply"""
    if n <= 1:
        return 1
    result = 1
    for i in range(1, n + 1):
        result = calc.multiply(result, i)
    return result

def complex_calculation(a, b, c, d):
    """Complex calculation: ((a + b) * c) - d"""
    step1 = calc.add(a, b)
    step2 = calc.multiply(step1, c)
    result = calc.subtract(step2, d)
    return result

# Add functions to calculator_app module
calculator_app.square = square
calculator_app.cube = cube
calculator_app.area_rectangle = area_rectangle
calculator_app.average = average
calculator_app.factorial = factorial
calculator_app.complex_calculation = complex_calculation

# Register in sys.modules
sys.modules['calculator_app'] = calculator_app

print("✓ calculator_app module created successfully!")
print("  Functions: square, cube, area_rectangle, average, factorial, complex_calculation")
print("  All functions utilize calc module functions!")
````

## Section 10: Using calculator_app Module with IMPORT

Let's use the calculator_app module with the `import` statement.

````python
# Using import style with calculator_app
import calculator_app

print("=== Using calculator_app with IMPORT style ===\n")

print("1. Square Numbers:")
for n in [3, 5, 7]:
    result = calculator_app.square(n)
    print(f"   square({n}) = {result}")

print("\n2. Cube Numbers:")
for n in [2, 3, 4]:
    result = calculator_app.cube(n)
    print(f"   cube({n}) = {result}")

print("\n3. Rectangle Area:")
areas = [(5, 10), (7, 4), (12, 8)]
for length, width in areas:
    result = calculator_app.area_rectangle(length, width)
    print(f"   area_rectangle({length}, {width}) = {result}")

print("\n4. Average of Numbers:")
lists = [[10, 20, 30], [5, 15, 25, 35], [100, 50]]
for lst in lists:
    result = calculator_app.average(lst)
    print(f"   average({lst}) = {result}")

print("\n5. Factorial:")
for n in [3, 5, 6]:
    result = calculator_app.factorial(n)
    print(f"   factorial({n}) = {result}")

print("\n6. Complex Calculation ((a+b)*c)-d:")
result1 = calculator_app.complex_calculation(10, 5, 3, 8)
result2 = calculator_app.complex_calculation(20, 10, 2, 5)
print(f"   ((10 + 5) * 3) - 8 = {result1}")
print(f"   ((20 + 10) * 2) - 5 = {result2}")
````

## Summary: Modules Overview

### What We Created:

#### 1. **calc Module** (Basic Math Functions)
Contains 7 functions:
- `add(x, y)` - Basic addition
- `subtract(x, y)` - Basic subtraction
- `multiply(x, y)` - Multiplication using repeated addition
- `divide_integer(x, y)` - Integer division using repeated subtraction
- `power(x, y)` - Power calculation using multiply
- `sum_list(numbers)` - Sum list items using add
- `difference_sum(n1, n2, n3)` - Combines add and subtract

#### 2. **calculator_app Module** (High-Level Operations)
Contains 6 functions that USE calc module:
- `square(x)` - Uses calc.power()
- `cube(x)` - Uses calc.power()
- `area_rectangle(l, w)` - Uses calc.multiply()
- `average(numbers)` - Uses calc.sum_list() and calc.divide_integer()
- `factorial(n)` - Uses calc.multiply()
- `complex_calculation(a, b, c, d)` - Uses calc.add(), calc.multiply(), calc.subtract()

### Import Methods Used:

```python
# Method 1: Import entire module
import calc
result = calc.add(5, 3)

# Method 2: Import with alias
import calc as c
result = c.add(5, 3)

# Method 3: From import (specific functions)
from calc import add, multiply
result = add(5, 3)
result = multiply(4, 5)

# Method 4: From import (multiple functions)
from calculator_app import square, cube, average
result = square(5)
result = cube(3)
result = average([10, 20, 30])
```

### Hierarchy:
```
calc module (base functions)
    ↓
calculator_app module (builds on calc module)
    └─ Uses calc.power(), calc.multiply(), calc.add(), calc.subtract(), calc.divide_integer()
```
