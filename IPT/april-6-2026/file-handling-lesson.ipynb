## Section 2: Understanding File Paths

A file path tells Python where to find a file. There are two types of paths:

### Absolute Paths
- Complete path from the root directory
- Example: `/Users/user/Documents/file.txt` (Mac/Linux) or `C:\Users\user\Documents\file.txt` (Windows)

### Relative Paths
- Path relative to the current working directory
- Example: `documents/file.txt`

**Why use pathlib?**
- Cross-platform compatibility (works on Windows, Mac, Linux)
- Less error-prone than string concatenation
- Object-oriented approach

````python
# Working with file paths using pathlib (RECOMMENDED)

# Create a Path object for current directory
current_dir = Path.cwd()
print(f"Current directory: {current_dir}")
print(f"Type: {type(current_dir)}")

# Create a path to a file
file_path = current_dir / "sample_file.txt"  # Use / operator to join paths
print(f"\nFile path: {file_path}")

# Get parts of the path
print(f"\nPath components:")
print(f"  Name: {file_path.name}")              # Just the filename
print(f"  Parent: {file_path.parent}")          # Parent directory
print(f"  Stem: {file_path.stem}")              # Filename without extension
print(f"  Suffix: {file_path.suffix}")          # File extension

# Check if path exists
print(f"\nPath exists? {file_path.exists()}")

# Create a directory path
folder_path = current_dir / "my_data"
print(f"\nFolder path: {folder_path}")
print(f"Folder exists? {folder_path.exists()}")
````

## Section 3: Opening and Closing Files

The `open()` function is used to open files in Python.

**Syntax:** `open(filename, mode)`

### File Modes:
| Mode | Meaning | Creates File? |
|------|---------|---|
| `'r'` | Read (default) | No |
| `'w'` | Write (overwrites) | Yes |
| `'a'` | Append (adds to end) | Yes |
| `'x'` | Create (fails if exists) | Yes |
| `'b'` | Binary mode | - |
| `'t'` | Text mode (default) | - |

**Common combinations:**
- `'rb'` - Read binary
- `'wb'` - Write binary
- `'ab'` - Append binary
- `'r+'` - Read and write

````python
# Example 1: Basic file opening and closing (NOT RECOMMENDED - can leave file open if error occurs)
print("=== Example 1: Manual open/close (NOT BEST PRACTICE) ===")

# Create a sample file first
sample_file = Path("sample.txt")
file = open(sample_file, 'w')
file.write("Hello, World!\nThis is Python file handling.\n")
file.close()  # Important: Always close the file!

print(f"File created: {sample_file.exists()}")

# Now read it back
file = open(sample_file, 'r')
content = file.read()
print(f"Content read: {content}")
file.close()

print(f"File closed? {file.closed}")  # Check if file is closed
````

## Section 4: Reading Files

There are several ways to read file contents in Python. Each method is useful in different situations.

### Reading Methods:
1. **`read()`** - Reads entire file as one string
2. **`readline()`** - Reads one line at a time
3. **`readlines()`** - Reads all lines as a list
4. **Iteration** - Loop through file line by line

````python
# Create a sample file with multiple lines for demonstration
sample_file = Path("multiline_sample.txt")
with open(sample_file, 'w') as f:
    f.write("Line 1: Introduction to file handling\n")
    f.write("Line 2: Python makes it easy\n")
    f.write("Line 3: Always close your files\n")
    f.write("Line 4: Or use 'with' statement\n")

print("=== Method 1: read() - Reads entire file as string ===")
with open(sample_file, 'r') as f:
    content = f.read()
    print(f"Type: {type(content)}")
    print(f"Content:\n{content}")

print("\n=== Method 2: readline() - Reads one line at a time ===")
with open(sample_file, 'r') as f:
    line1 = f.readline()  # Read first line
    line2 = f.readline()  # Read second line
    print(f"Line 1: {line1.strip()}")
    print(f"Line 2: {line2.strip()}")

print("\n=== Method 3: readlines() - Reads all lines as list ===")
with open(sample_file, 'r') as f:
    lines = f.readlines()
    print(f"Type: {type(lines)}")
    print(f"Number of lines: {len(lines)}")
    for i, line in enumerate(lines, 1):
        print(f"  Line {i}: {line.strip()}")

print("\n=== Method 4: Iteration - Loop through file (MOST EFFICIENT) ===")
with open(sample_file, 'r') as f:
    for line_num, line in enumerate(f, 1):
        print(f"  Line {line_num}: {line.strip()}")
````

````python
# Handling large files efficiently
print("=== Reading Large Files Efficiently ===")

# Create a large file
large_file = Path("large_file.txt")
with open(large_file, 'w') as f:
    for i in range(1000):
        f.write(f"Line {i+1}: This is sample data\n")

print(f"Large file created with 1000 lines")

# GOOD: Read in chunks to save memory
print("\nReading file in chunks (MEMORY EFFICIENT):")
chunk_size = 1024  # Read 1KB at a time
with open(large_file, 'r') as f:
    chunk_num = 0
    while True:
        chunk = f.read(chunk_size)
        if not chunk:
            break
        chunk_num += 1
    print(f"  File read in {chunk_num} chunks")

# BETTER: Iterate line by line (Most efficient for text files)
print("\nIterating line by line (BEST FOR TEXT FILES):")
line_count = 0
with open(large_file, 'r') as f:
    for line in f:
        line_count += 1
print(f"  Total lines: {line_count}")
````

## Section 5: Writing to Files

Writing to files allows you to save data. Be careful with write mode ('w') as it overwrites existing content!

### Methods:
- **`write(string)`** - Writes a single string
- **`writelines(list)`** - Writes multiple lines (doesn't add newlines automatically)
- **`print(text, file=f)`** - Use print function to write

````python
print("=== Writing to Files ===\n")

# Example 1: write() method
output_file = Path("output.txt")
with open(output_file, 'w') as f:
    f.write("First line of text\n")
    f.write("Second line of text\n")
    f.write("Third line of text\n")

print("Example 1: Using write() method")
with open(output_file, 'r') as f:
    print(f.read())

# Example 2: writelines() method
output_file2 = Path("output2.txt")
lines = ["Line 1: Python\n", "Line 2: File Handling\n", "Line 3: Easy!\n"]
with open(output_file2, 'w') as f:
    f.writelines(lines)

print("Example 2: Using writelines() method")
with open(output_file2, 'r') as f:
    print(f.read())

# Example 3: Using print() function
output_file3 = Path("output3.txt")
with open(output_file3, 'w') as f:
    print("Message 1", file=f)
    print("Message 2", file=f)
    print("Message 3", file=f)

print("Example 3: Using print() function")
with open(output_file3, 'r') as f:
    print(f.read())

# IMPORTANT: Difference between write modes
print("=== IMPORTANT: 'w' mode OVERWRITES existing content ===")
overwrite_file = Path("overwrite_demo.txt")

# First write
with open(overwrite_file, 'w') as f:
    f.write("Original content\n")

print("First write completed")
with open(overwrite_file, 'r') as f:
    print(f"Content: {f.read()}")

# Second write with 'w' mode - THIS OVERWRITES!
with open(overwrite_file, 'w') as f:
    f.write("New content (original is GONE!)\n")

print("\nAfter second write with 'w' mode:")
with open(overwrite_file, 'r') as f:
    print(f"Content: {f.read()}")
````

## Section 6: Appending to Files

Append mode ('a') adds content to the end of a file without deleting existing content. This is useful for log files and data collection.

````python
print("=== Appending to Files ===\n")

append_file = Path("append_demo.txt")

# First, write initial content
print("Step 1: Write initial content")
with open(append_file, 'w') as f:
    f.write("Original line 1\n")
    f.write("Original line 2\n")

with open(append_file, 'r') as f:
    print("Current content:")
    print(f.read())

# Now append more content
print("\nStep 2: Append new content")
with open(append_file, 'a') as f:
    f.write("Appended line 3\n")
    f.write("Appended line 4\n")

print("After appending:")
with open(append_file, 'r') as f:
    print(f.read())

# Practical example: Log file
print("\n=== PRACTICAL EXAMPLE: Simple Log File ===")
log_file = Path("app.log")

# Simulate logging over time
import datetime

def log_message(message):
    """Append a timestamped message to the log file"""
    timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(log_file, 'a') as f:
        f.write(f"[{timestamp}] {message}\n")

# Log multiple messages
log_message("Application started")
log_message("Processing data")
log_message("Data processed successfully")

print("\nLog file contents:")
with open(log_file, 'r') as f:
    print(f.read())
````

## Section 7: Working with Different File Formats

Python can work with various file formats. Each has its own module for easy handling.

### Formats We'll Cover:
1. **CSV (Comma-Separated Values)** - Data in tabular format
2. **JSON (JavaScript Object Notation)** - Structured data format
3. **Plain Text** - Simple text files

````python
print("=== Working with CSV Files ===\n")

# Create a CSV file
csv_file = Path("students.csv")

# Method 1: Write CSV manually (NOT RECOMMENDED)
print("Method 1: Manual CSV writing (NOT RECOMMENDED)")
with open(csv_file, 'w') as f:
    f.write("Name,Age,Grade\n")
    f.write("Alice,20,A\n")
    f.write("Bob,19,B\n")
    f.write("Charlie,21,A\n")

print("CSV file created (manual method):")
with open(csv_file, 'r') as f:
    print(f.read())

# Method 2: Using csv module (RECOMMENDED)
print("\nMethod 2: Using csv module (RECOMMENDED)")

# Write using csv.writer
csv_file2 = Path("students_proper.csv")
with open(csv_file2, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Name', 'Age', 'Grade'])
    writer.writerow(['Diana', 20, 'A'])
    writer.writerow(['Eve', 19, 'B+'])
    writer.writerow(['Frank', 21, 'A'])

# Read using csv.reader
print("Reading CSV with csv.reader:")
with open(csv_file2, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        print(f"  {row}")

# Method 3: Using csv.DictReader and DictWriter (BEST for structured data)
print("\nMethod 3: Using csv.DictReader/DictWriter (BEST)")

csv_file3 = Path("students_dict.csv")
students = [
    {'Name': 'Grace', 'Age': 20, 'Grade': 'A'},
    {'Name': 'Henry', 'Age': 19, 'Grade': 'B'},
    {'Name': 'Iris', 'Age': 21, 'Grade': 'A+'}
]

# Write
with open(csv_file3, 'w', newline='') as f:
    fieldnames = ['Name', 'Age', 'Grade']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(students)

# Read as dictionaries
print("Reading CSV as dictionaries:")
with open(csv_file3, 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"  {dict(row)}")
````

````python
print("=== Working with JSON Files ===\n")

# Create sample data
student_data = {
    'name': 'John Doe',
    'age': 20,
    'courses': ['Python', 'Web Development', 'Data Science'],
    'grades': {
        'Python': 95,
        'Web Development': 88,
        'Data Science': 92
    }
}

# Write to JSON
json_file = Path("student.json")
with open(json_file, 'w') as f:
    json.dump(student_data, f, indent=4)

print("JSON file created:")
with open(json_file, 'r') as f:
    print(f.read())

# Read from JSON
print("\nReading JSON file:")
with open(json_file, 'r') as f:
    loaded_data = json.load(f)
    print(f"Type: {type(loaded_data)}")
    print(f"Name: {loaded_data['name']}")
    print(f"Courses: {loaded_data['courses']}")
    print(f"Grade in Python: {loaded_data['grades']['Python']}")

# Write list of objects
json_file2 = Path("students.json")
students_list = [
    {'name': 'Alice', 'age': 20, 'major': 'CS'},
    {'name': 'Bob', 'age': 19, 'major': 'Math'},
    {'name': 'Charlie', 'age': 21, 'major': 'Physics'}
]

with open(json_file2, 'w') as f:
    json.dump(students_list, f, indent=2)

print("\n\nList of students in JSON:")
with open(json_file2, 'r') as f:
    print(f.read())
````

## Section 8: Error Handling in File Operations

File operations can fail for many reasons. Always handle errors gracefully!

### Common Errors:
- **`FileNotFoundError`** - File doesn't exist
- **`PermissionError`** - No permission to access file
- **`IOError`** - General input/output error
- **`IsADirectoryError`** - Trying to open a directory as file

````python
print("=== Error Handling in File Operations ===\n")

# Example 1: Handling FileNotFoundError
print("Example 1: Handling FileNotFoundError")
try:
    with open("nonexistent_file.txt", 'r') as f:
        content = f.read()
except FileNotFoundError:
    print("  ❌ Error: File not found!")
except Exception as e:
    print(f"  ❌ Unexpected error: {e}")
else:
    print(f"  ✓ File read successfully")

# Example 2: Handling multiple exceptions
print("\nExample 2: Handling multiple exceptions")
def safe_read_file(filename):
    try:
        with open(filename, 'r') as f:
            return f.read()
    except FileNotFoundError:
        return "Error: File not found"
    except PermissionError:
        return "Error: Permission denied"
    except IOError as e:
        return f"Error: IO problem - {e}"
    except Exception as e:
        return f"Error: Unexpected problem - {e}"

result = safe_read_file("another_nonexistent.txt")
print(f"  Result: {result}")

# Example 3: Try-except-else-finally
print("\nExample 3: Try-except-else-finally pattern")
test_file = Path("test_error_handling.txt")
test_file.write_text("Test content")

try:
    with open(test_file, 'r') as f:
        content = f.read()
except FileNotFoundError:
    print("  File not found")
except Exception as e:
    print(f"  Error occurred: {e}")
else:
    print(f"  ✓ File read successfully: {content.strip()}")
finally:
    print("  (Cleanup code runs regardless of what happened)")

# Example 4: Check if file exists before reading
print("\nExample 4: Check existence before reading")
file_to_check = Path("data.txt")

if file_to_check.exists():
    if file_to_check.is_file():
        with open(file_to_check, 'r') as f:
            print(f"  ✓ Content: {f.read()}")
    else:
        print("  It's a directory, not a file!")
else:
    print("  File doesn't exist - creating it...")
    file_to_check.write_text("New file content")
````

## Section 9: Context Managers (with statement) - BEST PRACTICE

The `with` statement is the **recommended way** to handle files. It automatically closes files even if an error occurs!

**Why use `with`?**
- Automatically closes files
- Handles exceptions properly
- Cleaner, more readable code
- Prevents resource leaks

````python
print("=== Context Managers (with statement) ===\n")

# Create test file
test_file = Path("context_test.txt")
test_file.write_text("Line 1\nLine 2\nLine 3\n")

# Example 1: Without 'with' statement (BAD - file might not close if error occurs)
print("Example 1: WITHOUT 'with' statement (NOT RECOMMENDED)")
f = open(test_file, 'r')
try:
    # If an error occurs here, file won't close properly!
    content = f.read()
    print(f"  Content: {content.strip()}")
finally:
    f.close()
    print(f"  File closed? {f.closed}")

# Example 2: With 'with' statement (GOOD - automatically closes)
print("\nExample 2: WITH 'with' statement (RECOMMENDED)")
with open(test_file, 'r') as f:
    content = f.read()
    print(f"  Content: {content.strip()}")
    print(f"  Inside 'with': File open? {not f.closed}")

print(f"  Outside 'with': File closed? {f.closed}")

# Example 3: Multiple file operations with 'with'
print("\nExample 3: Reading from one file and writing to another")
input_file = Path("input_context.txt")
output_file = Path("output_context.txt")

input_file.write_text("Original line 1\nOriginal line 2\nOriginal line 3\n")

with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
    for line_num, line in enumerate(infile, 1):
        # Add line numbers while copying
        outfile.write(f"{line_num}: {line}")

print("  Files processed successfully!")
print("  Output file contents:")
with open(output_file, 'r') as f:
    print(f.read())

# Example 4: Exception safety with 'with'
print("\nExample 4: Exception safety with 'with' statement")
problem_file = Path("problem_file.txt")
problem_file.write_text("Some data")

try:
    with open(problem_file, 'r') as f:
        print(f"  File opened: {not f.closed}")
        # Simulate an error
        raise ValueError("Something went wrong!")
except ValueError:
    print(f"  Error caught, but file is safely closed: {f.closed}")
    print("  The 'with' statement handled file closing automatically!")
````

## Section 10: File Methods and Attributes

File objects have useful methods and attributes for advanced file operations.

### Important Methods:
- **`tell()`** - Returns current position in file
- **`seek()`** - Move to specific position
- **`flush()`** - Write buffered data to disk
- **`truncate()`** - Resize file to specific size

### Important Attributes:
- **`name`** - File name
- **`mode`** - File opening mode
- **`closed`** - Whether file is closed

````python
print("=== File Methods and Attributes ===\n")

# Create a test file
methods_file = Path("methods_test.txt")
methods_file.write_text("0123456789ABCDEFGHIJ")

# Example 1: File attributes
print("Example 1: File Attributes")
with open(methods_file, 'r') as f:
    print(f"  Name: {f.name}")
    print(f"  Mode: {f.mode}")
    print(f"  Closed: {f.closed}")
    print(f"  Type: {type(f)}")

# Example 2: tell() and seek() - Moving position
print("\nExample 2: tell() and seek() - File Position")
with open(methods_file, 'r') as f:
    print(f"  Initial position: {f.tell()}")
    
    first_char = f.read(1)
    print(f"  After reading 1 char ('{first_char}'): position = {f.tell()}")
    
    second_char = f.read(1)
    print(f"  After reading 1 more char ('{second_char}'): position = {f.tell()}")
    
    # Go back to the beginning
    f.seek(0)
    print(f"  After seek(0): position = {f.tell()}")
    
    # Read from beginning
    first_five = f.read(5)
    print(f"  Reading 5 chars from start: '{first_five}'")
    
    # Jump to position 10
    f.seek(10)
    print(f"  After seek(10): position = {f.tell()}")
    rest = f.read()
    print(f"  Reading from position 10: '{rest}'")

# Example 3: Seeking from different positions
print("\nExample 3: Seeking from different positions")
large_file = Path("large_seek.txt")
large_file.write_text("A" * 50)

with open(large_file, 'r') as f:
    # Position from start (0)
    f.seek(5, 0)  # 5 bytes from start
    print(f"  seek(5, 0): position = {f.tell()}")
    
    # Position relative to current (1)
    f.seek(10, 1)  # 10 bytes from current position
    print(f"  After seek(10, 1): position = {f.tell()}")
    
    # Position from end (2)
    f.seek(-5, 2)  # 5 bytes before end
    print(f"  After seek(-5, 2): position = {f.tell()}")
    last_five = f.read()
    print(f"  Last 5 characters: '{last_five}'")

# Example 4: truncate() - Resize file
print("\nExample 4: truncate() - Resizing file")
truncate_file = Path("truncate_test.txt")
truncate_file.write_text("This is a long line that we'll truncate")

print(f"  Original size: {truncate_file.stat().st_size} bytes")

with open(truncate_file, 'r+') as f:
    f.truncate(10)  # Keep only first 10 bytes

print(f"  After truncate(10): {truncate_file.stat().st_size} bytes")
print(f"  New content: '{truncate_file.read_text()}'")

# Example 5: flush() - Write to disk immediately
print("\nExample 5: flush() - Immediate write to disk")
flush_file = Path("flush_test.txt")

with open(flush_file, 'w') as f:
    f.write("Data before flush")
    print(f"  Data written but not flushed to disk")
    f.flush()  # Force write to disk
    print(f"  Data flushed to disk with flush()")
````

## Section 11: Working with Directories

Besides files, you often need to work with directories. The `os` and `pathlib` modules provide tools for this.

### Directory Operations:
- **Create directories** - `mkdir()`, `makedirs()`
- **List contents** - `listdir()`, `glob()`
- **Check existence** - `exists()`, `is_dir()`
- **Navigate** - `chdir()`, `getcwd()`
- **Remove** - `rmdir()`, `remove()`

````python
print("=== Working with Directories ===\n")

# Example 1: Create directories
print("Example 1: Creating Directories")
base_dir = Path("my_project")

# Create single directory
if not base_dir.exists():
    base_dir.mkdir()
    print(f"  Created: {base_dir}")

# Create nested directories
nested_dir = base_dir / "data" / "raw" / "input"
if not nested_dir.exists():
    nested_dir.mkdir(parents=True, exist_ok=True)
    print(f"  Created nested: {nested_dir}")

# Example 2: List directory contents
print("\nExample 2: Listing Directory Contents")
# Create some test files
(base_dir / "file1.txt").write_text("File 1")
(base_dir / "file2.txt").write_text("File 2")
(base_dir / "config.json").write_text("{}")

# Using os.listdir()
print("  Using os.listdir():")
for item in os.listdir(base_dir):
    full_path = base_dir / item
    item_type = "DIR" if full_path.is_dir() else "FILE"
    print(f"    {item_type}: {item}")

# Using pathlib (better)
print("\n  Using pathlib (BETTER):")
for item in base_dir.iterdir():
    item_type = "DIR" if item.is_dir() else "FILE"
    print(f"    {item_type}: {item.name}")

# Example 3: Glob patterns
print("\nExample 3: Finding Files with glob()")
txt_files = list(base_dir.glob("*.txt"))
print(f"  Text files in {base_dir}:")
for f in txt_files:
    print(f"    {f.name}")

# Recursive glob
all_txt_files = list(base_dir.glob("**/*.txt"))
print(f"\n  All .txt files (recursive):")
for f in all_txt_files:
    print(f"    {f}")

# Example 4: Check file/directory properties
print("\nExample 4: Checking File/Directory Properties")
print(f"  Is {base_dir} a directory? {base_dir.is_dir()}")
print(f"  Is {base_dir} a file? {base_dir.is_file()}")

test_file = base_dir / "file1.txt"
print(f"\n  Is {test_file.name} a directory? {test_file.is_dir()}")
print(f"  Is {test_file.name} a file? {test_file.is_file()}")
print(f"  File size: {test_file.stat().st_size} bytes")

# Example 5: Walk through directory tree
print("\nExample 5: Walking Directory Tree")
print(f"  Directory structure of {base_dir}:")

def show_tree(path, prefix=""):
    """Display directory tree"""
    if not path.is_dir():
        return
    
    items = sorted(path.iterdir())
    for i, item in enumerate(items):
        is_last = i == len(items) - 1
        current_prefix = "└── " if is_last else "├── "
        print(f"{prefix}{current_prefix}{item.name}")
        
        if item.is_dir():
            next_prefix = prefix + ("    " if is_last else "│   ")
            show_tree(item, next_prefix)

show_tree(base_dir)

# Example 6: Delete directories and files
print("\nExample 6: Deleting Files and Directories")

# Delete a file
file_to_delete = base_dir / "file2.txt"
if file_to_delete.exists():
    file_to_delete.unlink()
    print(f"  Deleted file: {file_to_delete.name}")

# Delete empty directory
empty_dir = base_dir / "empty_folder"
empty_dir.mkdir(exist_ok=True)
if empty_dir.exists() and not any(empty_dir.iterdir()):
    empty_dir.rmdir()
    print(f"  Deleted empty directory: {empty_dir.name}")

# Note: To delete non-empty directories, use shutil.rmtree()
````

Let's apply what we've learned to real-world scenarios!

````python
print("=== Practical Example 1: Log File Processor ===\n")

# Create a sample log file
log_file = Path("application.log")
with open(log_file, 'w') as f:
    f.write("2024-01-15 10:30:00 INFO Application started\n")
    f.write("2024-01-15 10:30:05 ERROR Database connection failed\n")
    f.write("2024-01-15 10:30:10 INFO Retrying connection\n")
    f.write("2024-01-15 10:30:12 INFO Database connected\n")
    f.write("2024-01-15 10:30:15 WARNING Low memory detected\n")

# Process log file and extract errors/warnings
def process_logs(filename):
    """Extract errors and warnings from log file"""
    errors = []
    warnings = []
    
    with open(filename, 'r') as f:
        for line_no, line in enumerate(f, 1):
            if 'ERROR' in line:
                errors.append((line_no, line.strip()))
            elif 'WARNING' in line:
                warnings.append((line_no, line.strip()))
    
    return errors, warnings

errors, warnings = process_logs(log_file)

print("Errors found:")
for line_no, message in errors:
    print(f"  Line {line_no}: {message}")

print("\nWarnings found:")
for line_no, message in warnings:
    print(f"  Line {line_no}: {message}")

# Save report
report_file = Path("log_report.txt")
with open(report_file, 'w') as f:
    f.write("LOG REPORT\n")
    f.write("=" * 50 + "\n\n")
    f.write(f"Total Errors: {len(errors)}\n")
    f.write(f"Total Warnings: {len(warnings)}\n\n")
    
    f.write("ERRORS:\n")
    for line_no, msg in errors:
        f.write(f"  {msg}\n")
    
    f.write("\nWARNINGS:\n")
    for line_no, msg in warnings:
        f.write(f"  {msg}\n")

print(f"\nReport saved to: {report_file}")

print("\n" + "="*60)
print("=== Practical Example 2: Data File Manipulation ===\n")

# Create sample data file
data_file = Path("sales_data.csv")
with open(data_file, 'w') as f:
    f.write("Month,Product,Sales\n")
    f.write("January,Laptop,5000\n")
    f.write("January,Phone,3000\n")
    f.write("February,Laptop,6000\n")
    f.write("February,Phone,4000\n")
    f.write("March,Laptop,5500\n")
    f.write("March,Phone,3500\n")

# Analyze sales data
def analyze_sales(filename):
    """Analyze sales data from CSV"""
    product_totals = {}
    month_totals = {}
    
    with open(filename, 'r') as f:
        next(f)  # Skip header
        for line in f:
            month, product, sales = line.strip().split(',')
            sales = int(sales)
            
            # Calculate totals
            product_totals[product] = product_totals.get(product, 0) + sales
            month_totals[month] = month_totals.get(month, 0) + sales
    
    return product_totals, month_totals

product_totals, month_totals = analyze_sales(data_file)

print("Sales by Product:")
for product, total in product_totals.items():
    print(f"  {product}: ${total}")

print("\nSales by Month:")
for month, total in month_totals.items():
    print(f"  {month}: ${total}")

print("\n" + "="*60)
print("=== Practical Example 3: File Backup Script ===\n")

# Create source files
backup_dir = Path("backup_demo")
source_dir = backup_dir / "source"
dest_dir = backup_dir / "backup"

source_dir.mkdir(parents=True, exist_ok=True)

# Create some files to backup
(source_dir / "important_file.txt").write_text("Important data")
(source_dir / "config.json").write_text('{"version": "1.0"}')
(source_dir / "readme.md").write_text("# README")

def backup_files(source, destination):
    """Simple backup function - copy files to destination"""
    destination.mkdir(parents=True, exist_ok=True)
    
    backed_up = []
    for file_path in source.iterdir():
        if file_path.is_file():
            dest_file = destination / file_path.name
            with open(file_path, 'r') as f_in:
                content = f_in.read()
            with open(dest_file, 'w') as f_out:
                f_out.write(content)
            backed_up.append(file_path.name)
    
    return backed_up

backed_files = backup_files(source_dir, dest_dir)
print(f"Files backed up: {len(backed_files)}")
for filename in backed_files:
    print(f"  ✓ {filename}")

print("\n" + "="*60)
print("=== Practical Example 4: Batch File Processing ===\n")

# Create multiple files with data
batch_dir = Path("batch_processing")
batch_dir.mkdir(exist_ok=True)

for i in range(1, 4):
    file = batch_dir / f"data_{i}.txt"
    file.write_text(f"Data file {i}\nLine 1\nLine 2\nLine 3\n")

def process_batch_files(directory, pattern="*.txt"):
    """Process multiple files matching pattern"""
    results = []
    
    for file_path in directory.glob(pattern):
        line_count = 0
        char_count = 0
        
        with open(file_path, 'r') as f:
            for line in f:
                line_count += 1
                char_count += len(line)
        
        results.append({
            'filename': file_path.name,
            'lines': line_count,
            'characters': char_count
        })
    
    return results

results = process_batch_files(batch_dir)

print("Batch Processing Results:")
for result in results:
    print(f"  File: {result['filename']}")
    print(f"    Lines: {result['lines']}, Characters: {result['characters']}\n")
````

## Summary: File Handling Best Practices

### Key Takeaways:

1. **Always use `with` statement** - Automatically closes files
   ```python
   with open('file.txt', 'r') as f:
       content = f.read()
   ```

2. **Use `pathlib` for paths** - Cross-platform compatible
   ```python
   from pathlib import Path
   file = Path('data') / 'file.txt'
   ```

3. **Handle exceptions** - Always prepare for errors
   ```python
   try:
       with open('file.txt', 'r') as f:
           content = f.read()
   except FileNotFoundError:
       print("File not found!")
   ```

4. **Choose the right read method**:
   - Small files: `read()`
   - Large files: Iterate line by line
   - Need lines as list: `readlines()`

5. **Use specialized modules**:
   - CSV files: `csv` module
   - JSON files: `json` module
   - Text processing: Built-in methods

6. **Check file existence** before operations
   ```python
   if Path('file.txt').exists():
       # Process file
   ```

7. **Use proper file modes**:
   - `'r'` for reading
   - `'w'` for writing (overwrites!)
   - `'a'` for appending
   - `'x'` for creating new files

8. **Close resources properly** - Let Python manage it with `with`

### Quick Reference:

| Task | Method |
|------|--------|
| Read whole file | `f.read()` |
| Read line by line | `for line in f:` |
| Write text | `f.write(text)` |
| Append to file | Open with mode `'a'` |
| Check if exists | `Path('file').exists()` |
| Create directory | `Path('dir').mkdir()` |
| List files | `list(Path('dir').glob('*.txt'))` |
| Get file size | `Path('file').stat().st_size` |
