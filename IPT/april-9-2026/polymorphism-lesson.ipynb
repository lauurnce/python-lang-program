## Summary

**Polymorphism** is powerful because it:
- Allows you to write flexible, reusable code
- Enables treating different objects uniformly through a common interface
- Supports three main forms in Python:
  - **Method Overriding**: Child classes override parent methods
  - **Duck Typing**: Objects are used based on their behavior, not their type
  - **Operator Overloading**: Custom behavior for built-in operators

With polymorphism, you can write functions that work with any object that has the required methods, making your code more maintainable and extensible!

```python
# Custom Vector class
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    # Overload the + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    # Overload the string representation
    def __str__(self):
        return f"Vector({self.x}, {self.y})"

# Use the overloaded operators
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2  # Uses __add__ method

print(f"{v1} + {v2} = {v3}")
```

## Bonus: Operator Overloading (Polymorphism with Operators)

You can redefine how operators like `+`, `-`, etc. work with custom objects using special methods.

```python
# Universal function that works with any object that has speak()
def make_it_speak(entity):
    print(entity.speak())

# This function works with all objects - Animals, Robot, Person
print("--- Using polymorphic function ---")
make_it_speak(dog)
make_it_speak(cat)
make_it_speak(robot)
make_it_speak(person)

# Can even store different types in a list and call them polymorphically
print("\n--- Polymorphic list ---")
all_entities = [dog, cat, cow, robot, person]
for entity in all_entities:
    make_it_speak(entity)
```

## 6. Polymorphism with Functions

A function that accepts any object with a `speak()` method, demonstrating true polymorphic behavior.

```python
# Classes that don't inherit from Animal but have speak method
class Robot:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return f"{self.name} says: Beep boop!"

class Person:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return f"{self.name} says: Hello!"

# Create instances
robot = Robot("T-800")
person = Person("Alice")

# Use them like animals - duck typing in action
print(robot.speak())
print(person.speak())
```

## 5. Demonstrate Duck Typing

"If it quacks like a duck, it's a duck." Objects don't need to inherit from the same base class to be used polymorphically.

```python
# Create instances of different animal types
dog = Dog("Rex")
cat = Cat("Whiskers")
cow = Cow("Bessie")

# Call the same method on different objects
# Different behavior for each class
print(dog.speak())
print(cat.speak())
print(cow.speak())
```

## 4. Demonstrate Method Overriding

This is the core of polymorphism. Same method name, different behavior based on the object type.

```python
# Child class 1: Dog
class Dog(Animal):
    # Override speak method
    def speak(self):
        return f"{self.name} says: Woof!"

# Child class 2: Cat
class Cat(Animal):
    # Override speak method
    def speak(self):
        return f"{self.name} says: Meow!"

# Child class 3: Cow
class Cow(Animal):
    # Override speak method
    def speak(self):
        return f"{self.name} says: Moo!"
```

# Polymorphism in Python

Polymorphism means "many forms." In Python, it allows objects of different types to be treated as objects of a common parent type. It's a key concept in Object-Oriented Programming that enables flexible and reusable code.

## Key Concepts:
- **Method Overriding**: Child classes provide their own implementation of a method defined in the parent class
- **Duck Typing**: If it walks like a duck and quacks like a duck, it's a duck (behavior-based)
- **Operator Overloading**: Using operators like `+`, `-` with custom objects
